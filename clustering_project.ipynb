<center><h1>Analyse de donn√©es CRM</h1></center>
<hr style="border-width:4px;border-color:#75DFC1">

>Une base de donn√©es CRM (Customer Relationship Management) est une collection organis√©e de donn√©es relatives aux interactions et aux comportements des clients d'une entreprise. Ces bases de donn√©es contiennent souvent des informations sur les transactions, les pr√©f√©rences d'achat, les interactions avec l'entreprise (par le biais de diff√©rents canaux, tels que les e-mails, les appels t√©l√©phoniques, les achats en ligne et en magasin), et parfois des donn√©es sociod√©mographiques comme l'√¢ge, le sexe, ou la localisation.

>L'objectif principal d'une base de donn√©es CRM est de centraliser toutes ces informations afin d'am√©liorer la gestion des relations avec les clients et de soutenir la prise de d√©cision dans le cadre des actions marketing, des ventes, du service client et de la fid√©lisation.


>
><h2 style="color:#75DFC1; font-family: 'Arial', sans-serif;">üîç Cadres d'utilisation d'une base de donn√©es CRM</h2>
> Les bases de donn√©es CRM sont utilis√©es dans divers contextes au sein des entreprises pour :

> 1. **Segmentation des clients :** Diviser la base de clients en groupes homog√®nes selon des crit√®res tels que la fr√©quence d'achat, le montant des d√©penses ou les cat√©gories de produits achet√©es. Cela permet d'adapter les actions marketing √† des segments sp√©cifiques (ex : clients VIP, nouveaux clients, clients dormants).

> 2. **Analyse de la valeur client (CLV) :** Estimer la valeur future des clients, souvent bas√©e sur leur historique d'achats, afin de prioriser les efforts de marketing et de fid√©lisation.

> 3. **Personnalisation des offres :** Utiliser les donn√©es CRM pour proposer des offres et recommandations adapt√©es aux pr√©f√©rences et aux comportements individuels des clients, augmentant ainsi l'engagement et les taux de conversion.

> 4. **Pr√©diction du churn :** Identifier les clients susceptibles de quitter l'entreprise (churn) en analysant les tendances d'interaction et d'achat, permettant d'intervenir avant qu'ils ne partent.

> 5. **Optimisation des campagnes marketing :** Planifier et ex√©cuter des campagnes marketing cibl√©es en fonction des pr√©f√©rences des clients ou de leur engagement pass√©, avec des promotions adapt√©es pour chaque segment.

>
><h2 style="color:#75DFC1; font-family: 'Arial', sans-serif;">üéØ Objectifs Cl√©s du Projet</h2>
> Dans notre projet, nous allons nous concentrer sur une approche de ciblage et de segmentation bas√©e sur les donn√©es CRM disponibles. Notre objectif principal est d'identifier des groupes distincts au sein de la base de donn√©es clients et de comprendre les caract√©ristiques qui les diff√©rencient.

>1. **Identification des segments de clients :** Nous allons utiliser des m√©thodes de segmentation, telles que le clustering, pour d√©terminer s'il existe des groupes homog√®nes au sein de notre base de donn√©es. Ces segments pourraient repr√©senter diff√©rents types de clients avec des comportements ou des caract√©ristiques similaires.

>2. **Analyse des caract√©ristiques des segments :** Une fois les segments identifi√©s, nous allons chercher √† comprendre ce qui distingue chaque groupe. Par exemple, certains segments pourraient regrouper des clients √† forte valeur, tandis que d'autres incluraient des clients qui privil√©gient un canal d'achat sp√©cifique.

>3. **Analyse statistique pr√©alable :** Avant de proc√©der √† la segmentation, nous allons v√©rifier si le clustering est appropri√© pour notre jeu de donn√©es. Pour ce faire, nous utiliserons des tests statistiques tels que les tests d'homog√©n√©it√©, la distribution des variables, et l'analyse des tendances. Ces tests nous indiqueront si les donn√©es pr√©sentent des tendances naturelles de regroupement qui justifieraient l'utilisation d'algorithmes de clustering.

```python
# Utilisation d'un contexte de style Seaborn temporaire pour cette cellule uniquement
with sns.axes_style("whitegrid"):
    # Cr√©ation d'une figure avec 2 sous-graphiques (1 ligne, 2 colonnes)
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 ligne, 2 colonnes

    # Liste des colonnes √† visualiser
    colonnes = ['order_channel', 'last_order_channel']

    # Parcours des colonnes et cr√©ation des histogrammes tri√©s
    for i, col in enumerate(colonnes):
        ax = axes[i]  # Utiliser l'axe correspondant
        
        # Calculer les valeurs de fr√©quence tri√©es pour chaque colonne
        sorted_order = Data_CRM[col].value_counts().index
        
        total = float(len(Data_CRM[col]))
        
        # Tracer les courbes avec les cat√©gories tri√©es
        sns.countplot(x=col, data=Data_CRM, palette="viridis", order=sorted_order, ax=ax)
        ax.set_title(f'Distribution de {col}')
        
        # Ajout des pourcentages sur les barres
        for p in ax.patches:
            height = p.get_height()
            ax.text(p.get_x() + p.get_width()/2.,
                    height + 3,
                    '{:1.2f}%'.format(100 * height / total),
                    ha="center")

    # Ajustement de l'espace entre les graphiques
    plt.tight_layout()
    plt.show()
```

Cette diff√©rence sugg√®re une distribution asym√©trique, probablement influenc√©e par des valeurs extr√™mes. Dans de telles situations, la m√©diane est souvent pr√©f√©r√©e √† la moyenne pour l'imputation car elle est moins sensible √† ces valeurs extr√™mes.

```python
# Calcul de la m√©diane pour la colonne 'Revenu'
revenu_median = Data_Campagne['Revenu'].median()

# Imputation des valeurs manquantes par la m√©diane
Data_Campagne['Revenu'].fillna(revenu_median, inplace=True)


# Calcul des statistiques descriptives pour 'Revenu'
stats_desc_revenu = Data_Campagne["Revenu"].describe().round(2)

# Convertir en DataFrame pour pouvoir transposer
stats_desc_revenu_df = pd.DataFrame(stats_desc_revenu)

# Transposition du tableau pour une meilleure pr√©sentation
stats_desc_revenu_transposed = stats_desc_revenu_df.T

# Affichage du tableau transpos√©
display(stats_desc_revenu_transposed)

```

    <h3 style="color: #75DFC1;">Discr√©tisation de la Variable 'Revenu'</h3>
    <p>
        La discr√©tisation de la variable 'Revenu' est une √©tape cl√© pour simplifier les analyses ult√©rieures et am√©liorer la performance des mod√®les pr√©dictifs. Nous utiliserons les quartiles pour diviser les revenus en quatre cat√©gories distinctes.
    </p>
    <p>
        En se basant sur les statistiques descriptives, nous d√©finirons les cat√©gories de revenu comme suit :
        <ul>
            <li><strong>Bas Revenu :</strong> Revenus inf√©rieurs au premier quartile (moins de 35,303 ‚Ç¨).</li>
            <li><strong>Revenu Moyen-Bas :</strong> Revenus entre le premier quartile et la m√©diane (35,303 ‚Ç¨ √† 51,381.5 ‚Ç¨).</li>
            <li><strong>Revenu Moyen-Haut :</strong> Revenus entre la m√©diane et le troisi√®me quartile (51,381.5 ‚Ç¨ √† 68,522 ‚Ç¨).</li>
            <li><strong>Haut Revenu :</strong> Revenus sup√©rieurs au troisi√®me quartile (plus de 68,522 ‚Ç¨).</li>
        </ul>
    </p>
    <p>
        Cette m√©thode de discr√©tisation nous permet de transformer les donn√©es continues du revenu en cat√©gories plus maniables, facilitant la compr√©hension des tendances et des patterns au sein de notre client√®le.
    </p>
</div>


```python
# D√©termination des quartiles
quartile1 = Data_Campagne['Revenu'].quantile(0.25)
quartile2 = Data_Campagne['Revenu'].quantile(0.5)  # M√©diane
quartile3 = Data_Campagne['Revenu'].quantile(0.75)

# Fonction pour cat√©goriser chaque revenu en classes nominales
def categorize_revenue(revenue):
    if revenue <= quartile1:
        return 'inf_35,303‚Ç¨'  # Revenus inf√©rieurs au premier quartile
    elif quartile1 < revenue <= quartile2:
        return '[35,303‚Ç¨_et_51,381.5‚Ç¨['  # Revenus entre le premier quartile et la m√©diane
    elif quartile2 < revenue <= quartile3:
        return '[51,381.5‚Ç¨_et_68,522‚Ç¨['  # Revenus entre la m√©diane et le troisi√®me quartile
    else:
        return 'sup_68,522‚Ç¨'  # Revenus sup√©rieurs au troisi√®me quartile

# Application de la fonction de cat√©gorisation au DataFrame
Data_Campagne['Categorie_Revenu'] = Data_Campagne['Revenu'].apply(categorize_revenue)

#Utilisation d'un contexte de style Seaborn temporaire pour cette cellule uniquement
with sns.axes_style("whitegrid"):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x='Categorie_Revenu', data=Data_Campagne, palette='viridis')
    ax.set_title('Distribution des Cat√©gories de Revenu')
    ax.set_xlabel('Cat√©gorie de Revenu')
    ax.set_ylabel('Nombre de Clients')

    # Calcul et affichage des pourcentages sur les barres
    total = len(Data_Campagne['Categorie_Revenu'])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

    plt.show()
```

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;">
    <h3 style="color: #75DFC1;">Calcul de l'√Çge et de l'Anciennet√© des Clients</h3>
    <p>
        Pour mieux comprendre notre client√®le, nous avons calcul√© l'√¢ge et l'anciennet√© de nos clients. L'√¢ge a √©t√© d√©duit √† partir de l'ann√©e de naissance, en consid√©rant 2019 comme l'ann√©e actuelle. L'anciennet√© a √©t√© calcul√©e √† partir de la date d'inscription, en prenant √©galement 2019 comme r√©f√©rence.

    <h3 style="color: #75DFC1;">Discr√©tisation de l'√Çge</h3>
    <p>
        Apr√®s avoir calcul√© l'√¢ge, nous avons proc√©d√© √† sa discr√©tisation en se basant sur les quartiles de la distribution des √¢ges. Cela nous a permis de diviser les √¢ges en quatre cat√©gories distinctes :
    </p>
    <ul>
        <li><strong>Inf√©rieur √† 42 ans :</strong> Regroupe les jeunes adultes.</li>
        <li><strong>Entre 42 et 49 ans :</strong> Cat√©gorise les adultes moyens.</li>
        <li><strong>Entre 49 et 60 ans :</strong> Correspond aux adultes sup√©rieurs.</li>
        <li><strong>Sup√©rieur √† 60 ans :</strong> Englobe les seniors.</li>
    </ul>
    <p>
        Cette approche de discr√©tisation nous aide √† mieux cerner les diff√©rents groupes d'√¢ge au sein de notre client√®le, facilitant ainsi l'analyse d√©mographique et la segmentation pour des strat√©gies marketing cibl√©es.

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;">
    <h3 style="color: #75DFC1;">Discr√©tisation de la R√©cence des Achats</h3>
    <p>
        La variable 'Recence_Achat' est un indicateur crucial de l'engagement des clients. Pour faciliter l'analyse, nous discr√©tisons cette variable en cat√©gories bas√©es sur la distribution des jours √©coul√©s depuis le dernier achat.
    </p>
    <ul>
        <li><strong>Tr√®s R√©cent :</strong> 0 √† 24 jours depuis le dernier achat.</li>
        <li><strong>R√©cent :</strong> 25 √† 49 jours depuis le dernier achat.</li>
        <li><strong>Moyennement Ancien :</strong> 50 √† 74 jours depuis le dernier achat.</li>
        <li><strong>Ancien :</strong> Plus de 74 jours depuis le dernier achat.</li>
    </ul>
    <p>
        Cette segmentation nous permet d'identifier des groupes de clients en fonction de la fr√©quence de leurs interactions, offrant des perspectives pr√©cieuses pour des strat√©gies marketing personnalis√©es.
    </p>
</div>


```python
def discretize_recency(value, quartiles):
    if value <= quartiles[0.25]:
        return "Tr√®s R√©cent"
    elif value <= quartiles[0.50]:
        return "R√©cent"
    elif value <= quartiles[0.75]:
        return "Moyennement Ancien"
    else:
        return "Ancien"

# Calcul des quartiles pour 'Recence_Achat'
quartiles_recency = Data_Campagne['Recence_Achat'].quantile([0.25, 0.50, 0.75])

# Application de la discr√©tisation
Data_Campagne['Categorie_Recence_Achat'] = Data_Campagne['Recence_Achat'].apply(lambda x: discretize_recency(x, quartiles_recency))

# Utilisation d'un contexte de style Seaborn temporaire pour cette cellule uniquement
with sns.axes_style("whitegrid"):
    # Cr√©ation de l'histogramme pour 'Categorie_Recence_Achat'
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x='Categorie_Recence_Achat', data=Data_Campagne, palette='viridis', order=['Tr√®s R√©cent', 'R√©cent', 'Moyennement Ancien', 'Ancien'])
    ax.set_title('Distribution des Cat√©gories de R√©cence des Achats')
    ax.set_xlabel('Cat√©gories de R√©cence')
    ax.set_ylabel('Nombre de Clients')

    # Calcul du total pour les pourcentages
    total = len(Data_Campagne['Categorie_Recence_Achat'])

    # Ajout des pourcentages sur les barres
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', fontsize=10, color='black', 
                    xytext=(0, 5), textcoords='offset points')

    plt.show()
```

    <h3 style="color: #75DFC1;">Introduction √† la R√©gression Logistique pour l'Analyse des Campagnes Marketing</h3>
    <p>
        La r√©gression logistique est un outil statistique puissant utilis√© pour mod√©liser la probabilit√© d'un √©v√©nement binaire ou cat√©goriel. Dans le cadre de notre projet, ce mod√®le sera utilis√© pour pr√©dire la probabilit√© de r√©ponse d'un client √† une campagne marketing (r√©ponse ou non-r√©ponse).
    </p>
    <p>
        <strong>Contexte du Mod√®le :</strong> La r√©gression logistique est particuli√®rement adapt√©e √† notre analyse car elle permet de traiter des probl√®mes de classification binaire, o√π l'objectif est de pr√©dire une variable d√©pendante binaire (dans notre cas, la 'Reponse' du client).
    </p>
    <p>
        <strong>√âquation Math√©matique :</strong> Le mod√®le s'articule autour de l'√©quation logistique, donn√©e par :
        <br>
        <i>P(Y=1) = 1 / (1 + e<sup>-(Œ≤<sub>0</sub> + Œ≤<sub>1</sub>X<sub>1</sub> + ... + Œ≤<sub>k</sub>X<sub>k</sub>)</sup>)</i>
        <br>
        o√π P(Y=1) est la probabilit√© que la variable d√©pendante (r√©ponse du client) soit √©gale √† 1, Œ≤<sub>i</sub> repr√©sente les coefficients du mod√®le, et X<sub>i</sub> sont les variables explicatives.
    </p>
    <p>
        <strong>Approche Mod√©lisatrice :</strong> Dans notre projet, nous adoptons une approche progressive, en commen√ßant par des mod√®les plus simples comme la r√©gression logistique. Cette m√©thode nous permet de comprendre les relations fondamentales dans les donn√©es avant de passer √† des mod√®les plus complexes. Cela facilite √©galement l'interpr√©tation des r√©sultats et la d√©termination des variables cl√©s influen√ßant la r√©ponse des clients.
    </p>
    <p>
        L'utilisation de la r√©gression logistique comme point de d√©part offre un √©quilibre entre simplicit√© et puissance pr√©dictive, permettant une premi√®re exploration efficace des donn√©es et des tendances sous-jacentes.
    </p>

# ****Mod√®le 3 : R√©gression Logistique avec ajustement pour le d√©s√©quilibre des classes, optimisation des hyperparam√®tres et sur√©chantillonnage SMOTE**** 

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;">
    <h3 style="color: #75DFC1;">Synth√®se des Am√©liorations Apport√©es au Mod√®le de Classification</h3>
    <p>
        Afin d'am√©liorer notre mod√®le de r√©gression logistique face √† un jeu de donn√©es pr√©sentant un d√©s√©quilibre de classes, nous avons d√©ploy√© une s√©rie de techniques avanc√©es d'analyse de donn√©es et de machine learning. Le processus d'optimisation a inclus l'utilisation de l'algorithme SMOTE pour le sur√©chantillonnage de la classe minoritaire, l'application de la standardisation des donn√©es pour normaliser les √©chelles des caract√©ristiques, et l'emploi d'une recherche al√©atoire d'hyperparam√®tres avec validation crois√©e stratifi√©e pour affiner notre mod√®le.
    </p>
    <p>
        La s√©lection des hyperparam√®tres a √©t√© guid√©e par le score AUC-ROC, privil√©giant ainsi la capacit√© du mod√®le √† distinguer les diff√©rentes classes. Une fois les meilleurs param√®tres identifi√©s, le mod√®le a √©t√© r√©√©valu√©, mettant en √©vidence une am√©lioration des mesures de performance, en particulier pour la classe minoritaire. 
    </p>
</div>


```python
# Configuration du pipeline avec sur√©chantillonnage SMOTE et standardisation
pipeline = ImbPipeline([
    ('smote', SMOTE(random_state=42)),
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])

# Grille d'hyperparam√®tres pour RandomizedSearchCV
param_distributions = {
    'classifier__C': [0.01, 0.1, 1, 10, 100],
    'classifier__solver': ['lbfgs', 'liblinear'],
    'classifier__class_weight': [None, 'balanced']
}

# Configuration de RandomizedSearchCV avec validation crois√©e stratifi√©e
random_search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_distributions,
    n_iter=20,  # Nombre d'it√©rations
    cv=StratifiedKFold(n_splits=5),  # Validation crois√©e stratifi√©e
    scoring='roc_auc',  # Utilisation de AUC-ROC comme m√©trique d'√©valuation
    random_state=42
)

# Entra√Ænement et recherche des meilleurs hyperparam√®tres
random_search.fit(X_train, y_train)

# R√©cup√©ration du meilleur mod√®le
best_model = random_search.best_estimator_

# Pr√©diction sur l'ensemble de test
y_pred_best = best_model.predict(X_test)
y_proba_best = best_model.predict_proba(X_test)[:, 1]

# √âvaluation du mod√®le
accuracy_best = accuracy_score(y_test, y_pred_best)
conf_matrix_best = confusion_matrix(y_test, y_pred_best)
class_report_best = classification_report(y_test, y_pred_best)
auc_score_best = roc_auc_score(y_test, y_proba_best)

# Affichage des r√©sultats
best_params = random_search.best_params_
best_params, 

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_best, annot=True, fmt='d', cmap='Blues', xticklabels=['Non R√©ponse', 'R√©ponse'], yticklabels=['Non R√©ponse', 'R√©ponse'])
plt.title('Matrice de Confusion - Mod√®le Am√©lior√©')
plt.xlabel('Pr√©dit')
plt.ylabel('R√©el')
plt.show()
print(accuracy_best)
print(class_report_best)
print(auc_score_best)
```
